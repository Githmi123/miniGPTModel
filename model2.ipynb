```python
import os
from openai import AsyncOpenAI

client = AsyncOpenAI(api_key='')
```

```python
prompts = [
    "Hi bro, how are you ah?",
    "Today very hot lah.",
    "Can you help me with this task meh?",
    "I don't know lor.",
    "Sure can one, no problem sia."
]

# Initialize an empty dataset list
dataset = []

# Loop through each prompt
for prompt in prompts:
    # Create a ChatCompletion using the gpt-3.5-turbo model
    response = await client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=50
    )
    
    # Extract the completion text from the response
    completion = response.choices[0].message['content'].strip()
    
    # Append the input-output pair to the dataset
    dataset.append({"input": prompt, "output": completion})

# Save the dataset to a JSONL file
with open("synthetic_singlish_dataset.jsonl", "w") as f:
    for entry in dataset:
        f.write(json.dumps(entry) + "\n")
```
