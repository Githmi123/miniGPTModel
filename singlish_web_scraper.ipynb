```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL of the webpage containing Singlish text
url = 'https://www.easyhindityping.com/sinhala-to-english-translation'

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract all text content from the webpage
    all_text = soup.get_text()

    # Define common Singlish terms to search for
    singlish_terms = ['lah', 'lor', 'leh', 'meh', 'lor', 'sia']

    # Find and print text containing any of the Singlish terms
    for line in all_text.split('\n'):
        if any(term in line.lower() for term in singlish_terms):
            print(line)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
